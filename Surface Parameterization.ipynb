# Surface

**What is a surface?** 

Before we define what a surface is, let us concretely define a "line". A line is geometrical object that can be parameterized using just one variable $t$. Below are all examples of lines. You just need to keep track of one variable to uniquely identify you on a line. In physics terms, you can also say that a line provides *one degree of freedom*.

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

# For circle
t = np.linspace(0, 2*np.pi, 120)
circleX, circleY = (np.cos(t), np.sin(t))

# For spiral
t = np.linspace(-2*np.pi, 2*np.pi, 120)
spiralX, spiralY, spiralZ = (np.cos(t), np.sin(t), t)

fig = plt.figure()
fig.suptitle("Examples of Lines")

ax = fig.add_subplot(1, 2, 1, aspect='equal')
ax.set_title("cos(t), sin(t)")
ax.plot(circleX, circleY)

ax2 = fig.add_subplot(1, 2, 2, aspect='equal', projection='3d')
ax2.set_title("cos(t), sin(t), t")
ax2.plot(spiralX, spiralY, spiralZ)
```

### Surface: A definition

We can easily picture various surfaces, both in 2 and 3-dimensions, and it shall be known to the reader that surfaces also exist in $n$-dimensions. The definition of a surface is a geometrical object that can be traversed using two parametric variables.
